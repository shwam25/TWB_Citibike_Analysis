```python
# Importing Dependencies
import pandas as pd
import csv
import time
import os
```

```python
#Defining source file path
july2019_csv = os.path.join("Resources","JC-201907-citibike-tripdata.csv")
august2019_csv = os.path.join("Resources","JC-201908-citibike-tripdata.csv")
september2019_csv = os.path.join("Resources","JC-201909-citibike-tripdata.csv")
october2019_csv = os.path.join("Resources","JC-201910-citibike-tripdata.csv")
november2019_csv = os.path.join("Resources","JC-201911-citibike-tripdata.csv")
december2019_csv = os.path.join("Resources","JC-201912-citibike-tripdata.csv")
january2020_csv = os.path.join("Resources","JC-202001-citibike-tripdata.csv")
february2020_csv = os.path.join("Resources","JC-202002-citibike-tripdata.csv")
march2020_csv = os.path.join("Resources","JC-202003-citibike-tripdata.csv")
april2020_csv = os.path.join("Resources","JC-202004-citibike-tripdata.csv")
may2020_csv = os.path.join("Resources","JC-202005-citibike-tripdata.csv")
june2020_csv = os.path.join("Resources","JC-202006-citibike-tripdata.csv")
```

```python
#Reading the files into Pandas
july2019 = pd.read_csv(july2019_csv)
august2019 = pd.read_csv(august2019_csv)
september2019 = pd.read_csv(september2019_csv)
october2019 = pd.read_csv(october2019_csv)                    
november2019 = pd.read_csv(november2019_csv)                    
december2019 = pd.read_csv(december2019_csv)                    
january2020 = pd.read_csv(january2020_csv)                    
february2020 = pd.read_csv(february2020_csv)                    
march2020 = pd.read_csv(march2020_csv)                    
april2020 = pd.read_csv(april2020_csv)                    
may2020 = pd.read_csv(may2020_csv)                    
june2020 = pd.read_csv(june2020_csv)
```

```python
july2019.name = 'july'
august2019.name = 'august'
september2019.name = 'september'
october2019.name = 'october'
november2019.name = 'november'
december2019.name = 'december'
```

```python
#Updating/adding columns 
dataframe_2019 = [july2019,august2019,september2019,october2019,november2019,december2019]

for df in dataframe_2019:
    df['age'] = 2019-df['birth year']
    df['month'] = df.name
    df['year'] = 2019
    newGenList = []
    genList = list(df['gender'])

    print(genList)
    for x in genList:
        if x == 1:
            newGenList.append('Male')
        elif x == 2:
            newGenList.append('Female')
        else:
            newGenList.append('Unknown')
    df['gender'] = ""
    df['gender'] = newGenList
```

```python
merge_2019_data = pd.concat(dataframe_2019)
merge_2019_data.head()
```

```python
#Updating/adding columns 
dataframe_2020 = [january2020,february2020,march2020,april2020,may2020,june2020]

for df in dataframe_2020:
    df['age'] = 2020-df['birth year']
    newGenList = []
    genList = list(df['gender'])
    for x in genList:
        if x == 1:
            newGenList.append('Male')
        elif x == 2:
            newGenList.append('Female')
        else:
            newGenList.append('Unknown')
    df['gender'] = ""
    df['gender'] = newGenList
```

```python
january2020['month'] = 'january'
february2020['month'] = 'february'
march2020['month'] = 'march'
april2020['month'] = 'april'
may2020['month'] = 'may'
june2020['month'] = 'june'

january2020['year'] = '2020'
february2020['year'] = '2020'
march2020['year'] = '2020'
april2020['year'] = '2020'
may2020['year'] = '2020'
june2020['year'] = '2020'
```

```python
merge_2020_data = pd.concat(dataframe_2020)
merge_2020_data.head()
```

```python
final_data = pd.concat([merge_2019_data,merge_2020_data],ignore_index=True)
```
